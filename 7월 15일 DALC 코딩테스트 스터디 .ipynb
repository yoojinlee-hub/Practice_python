**22.10 2의 거듭제곱 리스트 생성하기**

화면 캡처 2021-07-15 200825.png

```python
num1,num2=map(int,input().split()) #정수변환

li=[2**i for i in range(num1,num2+1)] #for i in range(start, stop + 1) 첫 번째 정수에서 두 번째 정수까지 숫자 생성(stop 포함)
del li[1]
del li[-2]

print(li)
```

**23.7 지뢰찾기**
KakaoTalk_20210715_205018561.jpg

```python
row, col = map(int, input().split())
matrix = []
for i in range(row):
    matrix.append(list(input()))

for i in range(row): #행
    for j in range(col): #열
        cnt = 0
        if matrix[i][j] == ".":
            for y in range(i-1, i+2):
                for x in range(j-1, j+2):
                    if not (y < 0 or x < 0 or y >= row or x >= col):
                        if matrix[y][x] == "*":
                            cnt += 1
            matrix[i][j] = cnt
            print(matrix[i][j], end="")
        else:
            print(matrix[i][j], end="")
    print()


#입력된 문자열이 2차원 리스트 형태로 들어감
#<input으로 한 줄씩 입력받은 뒤 list를 사용하면 문자열이 문자 리스트로 변환됩니다. 이걸 matrix에 append로 추가하면 2차원 리스트가 됩니다.>
```

**25.8 딕셔너리에서 특정 값 삭제하기**

딕셔너리.png

화면 캡처 2021-07-15 202836.png

```python
keys = input().split()
values = map(int, input().split())

x = dict(zip(keys, values))

x = {key: value for key,value in x.items() if (value !=30) and (key != 'delta')}

print(x)
```
