백준 10984번


https://www.acmicpc.net/problem/10984

```python
#학기의 수
T = int(input())

#학점 입력
for i in range(T):
    N = int(input())
    subjects=0
    grades=0
    
    for i in range(N):
      subject,grade = map(float,input().split()) 
      #입력 값을 변수 두개에 저장하기 - map을 사용하여 실수로 변환하기
      subjects += subject
      grades += subject*grade
      #각 학기에 대해 근우의 총 학점과 평점(GPA)을 출력한다. 정답과의 절대 오차는 10-1까지 허용한다.
    GPA = grades/subjects
      #숫자를 소수점 1째자리까지 출력.
    print(int(subjects),'%.1f'%GPA)
```

```python
def countdown(n):
  i = n # i = n+1
  def counter():
    nonlocal i #함수 바깥쪽의 지역변수 i를 변경할 수 있도록 함
    i -= 1
    return i+1 #그냥 i로 하니까 9876543210이 출력됨
  return counter

n = int(input())
 
c = countdown(n)
for i in range(n):
    print(c(), end=' ')
    #end=' ':한 칸씩 옆으로 출력
```

코딩도장 37.3 두 점 사이의 거리 구하기


```python
import math
 
class Point2D:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
 
length = 0.0
p = [Point2D(), Point2D(), Point2D(), Point2D()]
p[0].x, p[0].y, p[1].x, p[1].y, p[2].x, p[2].y, p[3].x, p[3].y = map(int, input().split())

#첫번째 점과 마지막 점의 거리
#a = abs(p[3].x - p[0].x)
#b = abs(p[3].y - p[0].y)
#length = math.sqrt(math.pow(a, 2) + math.pow(b, 2))

#단, 마지막 점은 다음 점이 없으므로 
#반복문은 for i in range(len(p) - 1):과 같이 만들어서 마지막 요소의 바로 앞 요소까지만 반복합니다. 
#만약 리스트의 요소 개수 len(p)에서 1을 빼지 않으면 p[i + 1]은 리스트의 범위를 벗어난 접근을 하게 되므로 주의해야 합니다.
for i in range(len(p) - 1):
  a = abs(p[i+1].x-p[i].x)
  b = abs(p[i+1].y-p[i].y)
  c = math.sqrt(math.pow(a, 2) + math.pow(b, 2)) #제곱해주는 함수 pow , 제곱근 구해주는 함수 sqrt
  length += c

print(length)
```

```python
#37.1 두 점 사이의 거리 구하기
import math
 
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
 
p1 = Point2D(x=30, y=20)    # 점1
p2 = Point2D(x=60, y=50)    # 점2
 
a = p2.x - p1.x    # 선 a의 길이
b = p2.y - p1.y    # 선 b의 길이
 
c = math.sqrt((a * a) + (b * b))    # (a * a) + (b * b)의 제곱근을 구함 //피타고라스의 정리
print(c)    # 42.42640687119285
```

```python
#37.2 사각형의 넓이 구하기
class Rectangle:
    def __init__(self, x1, y1, x2, y2):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
 
rect = Rectangle(x1=20, y1=20, x2=40, y2=30) #두 점을 이용한 사각형

#너비 width를 구하려면 rect.x2에서 rect.x1을 빼면 되고, 높이 height를 구하려면 rect.y2에서 rect.y1을 빼면 됩니다. 
#단, 음수가 나올 수도 있으므로 abs 함수를 사용하여 절댓값으로 만들어줍니다
width = abs(rect.x2 - rect.x1)
height = abs(rect.y2 - rect.y1)
area = width * height

print(area)
```
