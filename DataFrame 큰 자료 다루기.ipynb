```python
city_df['density'] = city_df['Population']/city_df['Land area (in sqKm)'] #인구밀도
city_df_high_density = city_df[city_df['density']>10000]
city_df_high_density.info()
```

```python
density_ranks=city_df.sort_values(by='density',ascending=False) #인구밀도 순위
```
