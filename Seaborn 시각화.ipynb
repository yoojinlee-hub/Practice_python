*카데고리별 시각화*

```python
laptops_df= pd.read_csv('/content/drive/MyDrive/laptops.csv')
laptops_df.head()
```

```python
laptops_df['os']
```

```python
laptops_df['os'].unique()
```

```python
sns.catplot(data=laptops_df,x='os',y='price',kind='box') #x축: 카테고리, y축: 보고싶은 값
```

```python
sns.catplot(data=laptops_df,x='os',y='price',kind='violin')
```

```python
sns.catplot(data=laptops_df,x='os',y='price',kind='strip') #값이 얼마나 많은지 알 수 있음
```

```python
laptops_df['processor_brand'].unique()
```

```python
sns.catplot(data=laptops_df,x='os',y='price',kind='strip',hue='processor_brand') #hue = 색
```

```python
sns.catplot(data=laptops_df,x='os',y='price',kind='swarm',hue='processor_brand') #점들이 겹쳐지지 않음
```
