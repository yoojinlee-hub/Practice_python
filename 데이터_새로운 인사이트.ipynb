**groupby**


```python
df = pd.read_csv('/content/drive/MyDrive/laptops.csv')
brand_nation = {
    'Dell':'U.S.',
    'Apple':'U.S.',
    'Acer':'Taiwan',
    'Lenovo':'China',
    'Alienware':'U.S.',
    'Microsoft':'U.S.',
    'Asus':'Taiwan'
}
df['brand_nation']=df['brand'].map(brand_nation)
df
```

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.max()
#pandas 업데이트 이후 사용불가
```

```python
nation_groups.mean()
```

```python
nation_groups.first()
```

```python
nation_groups.last()
```

```python
nation_groups.plot(kind='box',y='price')
```

```python
nation_groups.plot(kind='hist',y='price')
```

데이터 합치기 **merge**


```python
#innerJoin : 겹치는 부분 합치기
#Left Outer Join : 왼쪽만 합치기<오른쪽에 없더라도 합치기>
#Right Outer Join
#Full Outer Join: 양 쪽에 있는 모든 것을 합치기
```

```python
price_df = pd.read_csv('/content/drive/MyDrive/vegetable_price.csv')
quantitiy_df = pd.read_csv('/content/drive/MyDrive/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantitiy_df
```

```python
pd.merge(price_df,quantitiy_df,on='Product')
```

```python
pd.merge(price_df,quantitiy_df,on='Product',how='left')
```

```python
pd.merge(price_df,quantitiy_df,on='Product',how='right')
```

```python
pd.merge(price_df,quantitiy_df,on='Product',how='outer')
```
